```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Upload data set

```python
data = pd.read_csv('breast-cancer-wisconsin.csv', header=0, na_values=['?'])
data
```

# Summarize each column of data

```python
summary_data = data.describe()
summary_data
```

# Identify missing values

```python
missing_values = data.isna()
pd.set_option('display.max_rows', None)
missing_values
```

# Replace missing values with the mean of the column

```python
for column in data.columns:
    column_mean = data[column].mean(skipna=True)
    data[column].fillna(column_mean, inplace=True)
    
data
```

# Display the frequency table of the columns "Class" and "F6"

```python
freq_class_vs_f6 = pd.crosstab(data['Class'], data['F6'])
freq_class_vs_f6
```

# Display scatter plot of the columns from "F1" to "F6"

```python
sns.pairplot(data.iloc[:,1:7], kind='scatter')
plt.show()
```

# Display histograms of the columns from "F7" to "F9"

```python
fig, ax = plt.subplots()
data[['F7','F8','F9']].plot.hist(alpha=0.5, ax=ax)
ax.set_title('Histogram of F7 to F9')
ax.legend(['F7', 'F8', 'F9'])
plt.show()
```

# Display box plots of the columns from "F7" to "F9"

```python
data[['F7','F8','F9']].plot(kind='box')
plt.title('Box Plots of F7 to F9')
plt.show()
```
